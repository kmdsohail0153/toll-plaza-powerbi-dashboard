```python
# Replace 'toll_plaza_dataset.csv' with the actual filename
df = pd.read_csv('toll_plaza.csv')
```

```python
# Standardize column names
df.columns = df.columns.str.strip().str.upper().str.replace(' ', '_')

# Display the first few rows
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# prompt: remove all the null from the data

df_cleaned = df.dropna()
print("Shape of the dataframe after removing null values:", df_cleaned.shape)
print("\nMissing values after removal:")
print(df_cleaned.isnull().sum())
```

```python
# Drop rows with missing values
df.dropna(inplace=True)
```

```python
# Dataset information
df.info()

# Summary statistics
df.describe(include='all')
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='PLAZA_TYPE', order=df['PLAZA_TYPE'].value_counts().index)
plt.title('Distribution of Plaza Types')
plt.xlabel('Plaza Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
top_cities = df['PLAZA_CITY'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.title('Top 10 Cities with Most Plazas')
plt.xlabel('Number of Plazas')
plt.ylabel('City')
plt.show()
```

```python
state_counts = df['PLAZA_STATE'].value_counts()

plt.figure(figsize=(12,8))
sns.barplot(x=state_counts.values, y=state_counts.index)
plt.title('Number of Plazas by State')
plt.xlabel('Number of Plazas')
plt.ylabel('State')
plt.show()
```

```python
# Dropdown for State selection
state_dropdown = widgets.Dropdown(
    options=df['PLAZA_STATE'].unique(),
    description='State:',
    disabled=False,
)

# Dropdown for Plaza Type selection
plaza_type_dropdown = widgets.Dropdown(
    options=df['PLAZA_TYPE'].unique(),
    description='Plaza Type:',
    disabled=False,
)

# Function to filter and display data
def filter_data(state, plaza_type):
    filtered_df = df[(df['PLAZA_STATE'] == state) & (df['PLAZA_TYPE'] == plaza_type)]
    display(filtered_df)

# Interactive display
widgets.interact(filter_data, state=state_dropdown, plaza_type=plaza_type_dropdown)
```

```python
# Most common plaza type
df['PLAZA_TYPE'].value_counts()

# State with highest number of plazas
df['PLAZA_STATE'].value_counts()

# City with highest number of plazas
df['PLAZA_CITY'].value_counts()

# Concessionaire distribution
df['CONCESSIONAIRE'].value_counts()
```

```python
concession_state = df.groupby(['PLAZA_STATE', 'CONCESSIONAIRE']).size().unstack().fillna(0)
concession_state.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='tab20')
plt.title('Concessionaire Distribution Across States')
plt.xlabel('State')
plt.ylabel('Number of Plazas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
heat_data = df.groupby(['PLAZA_STATE', 'PLAZA_TYPE']).size().unstack().fillna(0)
plt.figure(figsize=(12, 8))
sns.heatmap(heat_data, annot=True, fmt='g', cmap='YlGnBu')
plt.title('Plaza Type Distribution by State')
plt.xlabel('Plaza Type')
plt.ylabel('State')
plt.show()
```

```python
plaza_counts = df['PLAZA_TYPE'].value_counts()
fig = px.pie(values=plaza_counts, names=plaza_counts.index, hole=0.4, title='Plaza Type Distribution')
fig.show()
```
